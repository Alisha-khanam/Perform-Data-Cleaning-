## **Missing values in age,cabin and embarked bold text columns.**

**More than 70% values are missing in cabin columns,will have to drop**

```python
df.isnull().sum()
```

```python
#drop the "Cabin" column from the dataframe
df=df.drop(columns='Cabin',axis=1)
```

```python
df_copy=df.copy()
```

```python
df_copy[df_copy['Age'].isnull()]
```

```python
df_copy['Age'].fillna(df_copy['Age'].mode()[0],inplace=True)
df_copy.isnull().sum()
```

```python
df_copy['Embarked'].fillna(df_copy['Embarked'].mode()[0],inplace=True)
df_copy.isnull().sum()
```

# **Outlier detection**

```python
#Handling outliers from fare column
#first quartile
Q1=np.percentile(df['Fare'],25)
Q3=np.percentile(df['Fare'],75)#third quartile

IQR=Q3-Q1

outlier_low=Q1-IQR*1.5
outlier_high=Q3*IQR*1.5

df=df[(df['Fare']>outlier_low)&(df['Fare']<outlier_high)]
df.shape
```

```python
#Handling outliers in age(almost normal)
df=df[df['Age']<(df['Age'].mean()+3*df['Age'].std())]
df.shape
```
